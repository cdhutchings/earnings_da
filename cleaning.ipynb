# Calling and Cleaning the Earnings Document

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
from numpy.random import randn, rand
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
xls = pd.ExcelFile("edited_table.xlsx")
```

```python
dframe = xls.parse("All")
```

```python
dframe.set_axis(["description","code","number_thou","median","perc_change_med","mean","perc_change_mean",
                 "10","20","25","30","40","60","70","75","80","90"],axis=1,inplace=True)
```

```python
dframe.head()
```

```python
table = dframe.loc[~dframe["mean"].isin(["x",np.nan])]
```

```python
red_table = pd.DataFrame(table.description.str.split(",",1).tolist(),columns=["type","location"])
```

```python
red_table.head()
```

```python
means = table["mean"]
means.head()
```

```python
red_table.merge(means, left_index=True,right_index=True, how="outer").sort_values("mean",ascending=False)
```
